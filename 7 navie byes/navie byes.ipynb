```python
# Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris, fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline

# For reproducibility
np.random.seed(42)

# -------------------------------
# 1. GaussianNB on Iris dataset (continuous features)
# -------------------------------
print("="*60)
print("Gaussian Naive Bayes on Iris Dataset")
print("="*60)

# Load iris dataset
iris = load_iris()
X_iris, y_iris = iris.data, iris.target

# Split into train/test
X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(
    X_iris, y_iris, test_size=0.3, random_state=42
)

# Train GaussianNB
gnb = GaussianNB()
gnb.fit(X_train_i, y_train_i)
y_pred_i = gnb.predict(X_test_i)
acc_gnb = accuracy_score(y_test_i, y_pred_i)
print(f"Accuracy: {acc_gnb:.3f}\n")

# -------------------------------
# 2. MultinomialNB on text data (word counts)
# -------------------------------
print("="*60)
print("Multinomial Naive Bayes on 20 Newsgroups (word counts)")
print("="*60)

# Load a subset of 20 newsgroups (two categories for simplicity)
categories = ['rec.sport.baseball', 'sci.space']
newsgroups = fetch_20newsgroups(subset='all', categories=categories, shuffle=True, random_state=42)

# Split into train/test
X_news, y_news = newsgroups.data, newsgroups.target
X_train_n, X_test_n, y_train_n, y_test_n = train_test_split(
    X_news, y_news, test_size=0.3, random_state=42
)

# Create a pipeline: CountVectorizer (word counts) + MultinomialNB
multinomial_pipe = make_pipeline(
    CountVectorizer(stop_words='english'),  # convert text to word count vectors
    MultinomialNB()
)
multinomial_pipe.fit(X_train_n, y_train_n)
y_pred_n = multinomial_pipe.predict(X_test_n)
acc_multinomial = accuracy_score(y_test_n, y_pred_n)
print(f"Accuracy: {acc_multinomial:.3f}\n")

# -------------------------------
# 3. BernoulliNB on same text data (binary word occurrence)
# -------------------------------
print("="*60)
print("Bernoulli Naive Bayes on 20 Newsgroups (binary word occurrence)")
print("="*60)

# Pipeline with binary=True CountVectorizer + BernoulliNB
bernoulli_pipe = make_pipeline(
    CountVectorizer(binary=True, stop_words='english'),  # binary features: 1 if word present, else 0
    BernoulliNB()
)
bernoulli_pipe.fit(X_train_n, y_train_n)
y_pred_b = bernoulli_pipe.predict(X_test_n)
acc_bernoulli = accuracy_score(y_test_n, y_pred_b)
print(f"Accuracy: {acc_bernoulli:.3f}\n")

# -------------------------------
# Summary and Explanation
# -------------------------------
print("="*60)
print("SUMMARY")
print("="*60)
print(f"GaussianNB (Iris, continuous)      : {acc_gnb:.3f}")
print(f"MultinomialNB (newsgroups, counts) : {acc_multinomial:.3f}")
print(f"BernoulliNB (newsgroups, binary)   : {acc_bernoulli:.3f}")
print("\n--- Which one is better for which? ---")
print("""
- GaussianNB: Best for continuous features that roughly follow a normal distribution.
               Example: Iris measurements (sepal length, petal width, etc.)

- MultinomialNB: Ideal for discrete count data, like word counts in documents.
                  It takes into account how many times a word appears.
                  Usually performs better when term frequency matters (e.g., longer documents).

- BernoulliNB: Designed for binary features (word present/absent).
                Good for short texts or when only the presence/absence of words is important.
                Sometimes works better than MultinomialNB on very short documents (like tweets).
""")
```
