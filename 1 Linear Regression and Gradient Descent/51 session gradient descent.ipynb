```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load real dataset (Boston housing)
data = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Boston.csv')

# Select one feature (RM = average number of rooms per house)
x = data['RM'].values
y = data['MEDV'].values

# Normalize (important for gradient descent stability)
x = (x - np.mean(x)) / np.std(x)
y = (y - np.mean(y)) / np.std(y)

# Initialize parameters
m = 0      # slope
b = 0      # intercept
alpha = 0.01   # learning rate
epochs = 1000  # number of iterations
n = len(x)

# To store cost values for each iteration
cost_list = []

# Gradient Descent Loop
for i in range(epochs):
    # Predicted values
    y_pred = m * x + b

    # Calculate gradients (partial derivatives)
    dm = (-2/n) * np.sum(x * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Update parameters
    m = m - alpha * dm
    b = b - alpha * db

    # Compute cost (Mean Squared Error)
    cost = np.mean((y - y_pred) ** 2)
    cost_list.append(cost)

# Print final parameters and cost
print(f"Final slope (m): {m}")
print(f"Final intercept (b): {b}")
print(f"Final cost: {cost}")

# Plot cost reduction
plt.plot(range(epochs), cost_list)
plt.xlabel('Iterations')
plt.ylabel('Cost (MSE)')
plt.title('Cost Reduction over Iterations')
plt.show()

# Plot the fitted line
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, m * x + b, color='red', label='Fitted Line')
plt.xlabel('Average Rooms (normalized)')
plt.ylabel('House Price (normalized)')
plt.legend()
plt.title('Linear Regression using Gradient Descent')
plt.show()
```

```python
import numpy as np

class GDRegressor:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.m = 0  # slope
        self.b = 0  # intercept

    def fit(self, X, y):
        n = len(X)

        for i in range(self.epochs):
            y_pred = self.m * X + self.b

            # gradients
            dm = (-2 / n) * np.sum(X * (y - y_pred))
            db = (-2 / n) * np.sum(y - y_pred)

            # update parameters
            self.m = self.m - self.learning_rate * dm
            self.b = self.b - self.learning_rate * db

    def predict(self, X):
        return self.m * X + self.b
```

```python
# Visual derivation of R formula
np.random.seed(42)
x = np.linspace(0, 10, 30)
y = 2 * x + 3 + np.random.normal(0, 2, 30)

# Fit model
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
y_pred = model.predict(x.reshape(-1, 1))
y_mean = np.mean(y)

# Calculate components
total_ss = np.sum((y - y_mean)**2)  # Total sum of squares
residual_ss = np.sum((y - y_pred)**2)  # Residual sum of squares
explained_ss = np.sum((y_pred - y_mean)**2)  # Explained sum of squares

r2 = 1 - (residual_ss / total_ss)

# Create visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Total variation (SST)
ax1 = axes[0, 0]
ax1.scatter(x, y, alpha=0.6, s=50)
ax1.axhline(y=y_mean, color='green', linestyle='--', linewidth=2, 
           label=f'Mean Line ( = {y_mean:.1f})')

# Draw lines from each point to mean
for xi, yi in zip(x, y):
    ax1.plot([xi, xi], [yi, y_mean], 'gray', alpha=0.3, linewidth=1)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('1. Total Variation (SST)\n(y - ) = How far data is from mean', 
              fontsize=14, fontweight='bold')
ax1.text(0.5, 0.95, f'SST = {total_ss:.1f}', transform=ax1.transAxes,
         fontsize=12, ha='center', bbox=dict(boxstyle="round,pad=0.3", 
                                              facecolor="white", alpha=0.8))
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Residual variation (SSE)
ax2 = axes[0, 1]
ax2.scatter(x, y, alpha=0.6, s=50)
ax2.plot(x, y_pred, 'r-', linewidth=2, label='Regression Line')

# Draw lines from each point to regression line
for xi, yi, ypi in zip(x, y, y_pred):
    ax2.plot([xi, xi], [yi, ypi], 'red', alpha=0.3, linewidth=1)

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('2. Unexplained Variation (SSE)\n(y - ) = Errors after using model', 
              fontsize=14, fontweight='bold')
ax2.text(0.5, 0.95, f'SSE = {residual_ss:.1f}', transform=ax2.transAxes,
         fontsize=12, ha='center', bbox=dict(boxstyle="round,pad=0.3", 
                                              facecolor="white", alpha=0.8))
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Explained variation (SSR)
ax3 = axes[1, 0]
ax3.scatter(x, y, alpha=0.3, s=30)  # Faint background
ax3.axhline(y=y_mean, color='green', linestyle='--', linewidth=2)
ax3.plot(x, y_pred, 'r-', linewidth=2, label='Regression Line')

# Draw lines from regression line to mean
for xi, ypi in zip(x, y_pred):
    ax3.plot([xi, xi], [y_mean, ypi], 'blue', alpha=0.3, linewidth=1)

ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_title('3. Explained Variation (SSR)\n( - ) = How much better than mean', 
              fontsize=14, fontweight='bold')
ax3.text(0.5, 0.95, f'SSR = {explained_ss:.1f}', transform=ax3.transAxes,
         fontsize=12, ha='center', bbox=dict(boxstyle="round,pad=0.3", 
                                              facecolor="white", alpha=0.8))
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: R calculation visualization
ax4 = axes[1, 1]
# Create a pie chart of variance decomposition
labels = ['Explained by Model', 'Unexplained (Error)']
sizes = [explained_ss, residual_ss]
colors = ['#4ECDC4', '#FF6B6B']
explode = (0.1, 0)

wedges, texts, autotexts = ax4.pie(sizes, explode=explode, labels=labels, colors=colors,
                                   autopct='%1.1f%%', shadow=True, startangle=90)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

ax4.set_title(f'4. R Calculation\nR = 1 - (SSE/SST) = 1 - ({residual_ss:.1f}/{total_ss:.1f}) = {r2:.3f}\n'
              f'Model explains {r2*100:.1f}% of variance', 
              fontsize=14, fontweight='bold')

# Add equation text
equation_text = ('R = 1 - SSE/SST\n'
                 '   = 1 - ' + r'$\frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2}$')
ax4.text(-1.5, -1.2, equation_text, fontsize=12, bbox=dict(boxstyle="round,pad=0.5", 
                                                            facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()

print("Formula Breakdown:")
print(f"Total Sum of Squares (SST) = (y - ) = {total_ss:.2f}")
print(f"Residual Sum of Squares (SSE) = (y - ) = {residual_ss:.2f}")
print(f"Explained Sum of Squares (SSR) = ( - ) = {explained_ss:.2f}")
print(f"R = 1 - (SSE/SST) = 1 - ({residual_ss:.2f}/{total_ss:.2f}) = {r2:.4f}")
print(f"Alternative: R = SSR/SST = {explained_ss:.2f}/{total_ss:.2f} = {explained_ss/total_ss:.4f}")
```
