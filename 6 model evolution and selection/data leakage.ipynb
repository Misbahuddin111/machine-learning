```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

np.random.seed(42)
n_samples = 100
X = np.random.randn(n_samples, 5)  # Random features (noise)
y = np.random.choice([0, 1], n_samples)  # Random binary target

# Create a leaky feature: directly based on y (simulating target leakage)
leaky_feature = y + np.random.randn(n_samples) * 0.1
X_leaky = np.hstack((X, leaky_feature.reshape(-1, 1)))  # Add leaky feature to X

# Train without leakage (using only original X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
acc_no_leak = accuracy_score(y_test, pred)
print(f"Accuracy without leakage: {acc_no_leak}")

# Train with leakage (using X_leaky)
X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(X_leaky, y, test_size=0.2)
model_l = LogisticRegression()
model_l.fit(X_train_l, y_train_l)
pred_l = model_l.predict(X_test_l)
acc_leak = accuracy_score(y_test_l, pred_l)
print(f"Accuracy with leakage: {acc_leak}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

np.random.seed(42)
n_samples = 1000
X = np.random.rand(n_samples, 10) * 100  # Features scaled up to make scaling impact clear
y = (np.sum(X, axis=1) > 500).astype(int)  # Binary target based on sum threshold

# With leakage: Scale entire dataset before splitting
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(X_scaled, y, test_size=0.2)
model_l = LogisticRegression()
model_l.fit(X_train_l, y_train_l)
pred_l = model_l.predict(X_test_l)
acc_leak = accuracy_score(y_test_l, pred_l)

# Without leakage: Split first, then scale
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # Note: transform only, no fit on test
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
pred = model.predict(X_test_scaled)
acc_no_leak = accuracy_score(y_test, pred)

print(f"Accuracy with leakage: {acc_leak}")
print(f"Accuracy without leakage: {acc_no_leak}")
```
